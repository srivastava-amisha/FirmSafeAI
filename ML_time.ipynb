```python
#################-----------------Multiple HPCs New Code --------------------------------------------------------------------------------

#change length of dataset, samplesperprogram, and the different time variations you want, program does rest!
filename = 'combined_ben-mal-timeseries.csv'                      # change file name
lengthofdataset = 440 			 	# length of the dataset (num samples)
samplesperprogram = 55					# how many samples per program
timeserieslength = 5					# Time series length you want
locofclasslabel = 30					# location of the class label (benign and malware) 
column_vals = [0]					   # which HPCs that should be utilized 


numsamples = samplesperprogram - (samplesperprogram % timeserieslength)
kyber1 = pd.read_csv(filename)       
frames = [kyber1]
mergedset = pd.concat(frames)
X = mergedset.iloc[:, column_vals].values
y = mergedset.iloc[:, locofclasslabel].values

X = np.reshape(X, (lengthofdataset*len(column_vals),1),order='F')
y = np.repeat(y, len(column_vals), axis = 0)


vpsplit_val = (((lengthofdataset/samplesperprogram) * math.floor(samplesperprogram/timeserieslength)) * len(column_vals))
temp = []
temp1 = []
temp2 = []
counter = 0
totalsamples = 0
val = 0

print(len(X))

#create individual time series arrays
while val < len(X):
	if counter < timeserieslength:
		temp1 = np.append(temp1,X[val])
		if counter == 0:
			temp2.append(y[val])
		counter +=1
		totalsamples += 1
		
	if counter == timeserieslength:
		counter = 0
		temp.append(temp1)
		temp1 = []

	if totalsamples == numsamples:
		val += samplesperprogram % timeserieslength
		totalsamples = 0
	val = val + 1


X = temp
y = temp2
X = np.array(X)
y = np.array(y)

X = np.vsplit(X,vpsplit_val)             
X = np.array(X)


print(len(X))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)



#this classifier works
tsf = TimeSeriesForestClassifier(n_estimators=10,min_interval=2)

tsf.fit(X_train, y_train)

y_pred = tsf.predict(X_test)
print("Number of testing samples", len(X_test))



print("Accuracy for time series length: ", timeserieslength , " HPC columns: " , column_vals,  ": ", np.mean(y_pred==y_test))
precision1 = precision_score(y_pred, y_test)
print('Precision: %f' % precision1)
recall1 = recall_score(y_pred, y_test)
print('Recall: %f' % recall1)

tp = 0
fp = 0
tn = 0
fn = 0
for i in range(0,len(y_test)):
  if y_test[i] == 0 and y_pred[i] == 0:
    tn += 1
  if y_test[i] == 1 and y_pred[i] == 1:
    tp += 1
  if y_test[i] == 0 and y_pred[i] == 1:
      fp += 1
  if y_test[i] == 1 and y_pred[i] == 0:
      fn += 1

print(tp)
print(fp)
print(tn)
print(fn)
print(tp/(tp + fp))
print(tp/(tp+fn))
for i in range(0,len(y_test)):
  if y_test[i] == 0 and y_pred[i] == 1:
    print('wrong good')
  if y_test[i] == 1 and y_pred[i] == 0:
    print('wrong bad')
#print(X_test)
print(y_pred)
print(y_test)
```
